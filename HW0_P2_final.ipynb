```python
#Problem 2.1
maxi = data.groupby('Species')['PetalLengthCm'].max()
print(maxi,"\n")
#Find all the maximum 'PetalLengthCm' for the 'Iris-versicolor'

get_species = data.groupby('Species')
#group the data by species
find_versi = get_species.get_group('Iris-versicolor')
#find all data for the 'Iris-versicolor' species
max_versi = find_versi.groupby('Species')['PetalLengthCm'].max()
# find the maximum 'PetalLengthCm' for the 'Iris-versicolor' species

print(f"\nAnswer: The maximal 'PetalLengthCm' of the 'Iris-versicolor' is:", max_versi[0])
#Answer: The maximal "PetalLengthCm" of the "Iris-versicolor" is 5.1
```

```python
print(data.groupby('Species')['PetalLengthCm'].min())
#Find all the minimum 'PetalLengthCm' for the 'Iris-versicolor'

min_versi = find_versi.groupby('Species')['PetalLengthCm'].min()
# find the minimum 'PetalLengthCm' for the 'Iris-versicolor' species

print(f"\nAnswer: The minimal 'PetalLengthCm' of the 'Iris-versicolor' is:",min_versi[0] )
#Answer: The minimal "PetalLengthCm" of the "Iris-versicolor" is 3.0
```

```python
# Problem 2.2
print(data.groupby('Species')['SepalWidthCm'].mean()) 
#showing all the averages of 'SepalWidthCm' categories via 'Species'

print('\n', data[['Species','SepalWidthCm']].groupby('Species').mean()) 
#(same information as above)

larg_avg = max(data.groupby('Species')['SepalWidthCm'].mean())
print(larg_avg)
#get the largest average of all the species

print(f"\nAnswer: The largest average 'SepalWidthCm' among all of the 3 species is the Iris-setosa with an average of:", larg_avg)
#Answer: The largest average "SepalWidthCm" among all of the 3 species, the Iris-setosa has the largest average of 3.418 and
```
